```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid", palette="muted")

file_path = "data/Nairobi Kenya 2024-05-14 to 2025-05-14.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
print("Dataset shape:", df.shape)

print("\nMissing values per column:")
print(df.isnull().sum())

df['datetime'] = pd.to_datetime(df['datetime'])

df = df.sort_values('datetime')

columns_to_keep = [
    'datetime', 'tempmax', 'tempmin', 'temp', 'humidity', 
    'precip', 'windspeed', 'conditions'
]
df = df[columns_to_keep]

df.head()
```

```python
df.describe()[['tempmax', 'tempmin', 'temp']]

plt.figure(figsize=(12,6))
plt.plot(df['datetime'], df['tempmax'], label='Max Temp', linewidth=2)
plt.plot(df['datetime'], df['tempmin'], label='Min Temp', linewidth=2, alpha=0.7)

plt.title("🌤️ Nairobi Temperature Trends (2024–2025)", fontsize=16, weight='bold', color="#5C3317")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['datetime'], df['tempmax'], color='#d4af37', label='Max Temp', linewidth=2.5)  # gold
plt.plot(df['datetime'], df['tempmin'], color='#5C3317', label='Min Temp', linewidth=2.5, linestyle='--')  # deep brown

plt.title("🌞 Nairobi Temperature Trends (May 2024 – May 2025)", fontsize=17, weight='bold', color='#5C3317', pad=15)
plt.xlabel("Date", fontsize=12, color='#5C3317')
plt.ylabel("Temperature (°C)", fontsize=12, color='#5C3317')
plt.legend(frameon=False, fontsize=11)
plt.grid(True, linestyle='--', alpha=0.3)

sns.despine(top=True, right=True)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(
    data=df, x='datetime', y='precip',
    color='#1e6091', label='Rainfall (mm)', linewidth=2.5
)

sns.lineplot(
    data=df, x='datetime', y='humidity',
    color='#76c893', label='Humidity (%)', linewidth=2.5, alpha=0.8
)

plt.title("🌧️ Nairobi Rainfall & Humidity Trends (May 2024 – May 2025)", fontsize=17, weight='bold', color='#5C3317', pad=15)
plt.xlabel("Date", fontsize=12, color='#5C3317')
plt.ylabel("Value", fontsize=12, color='#5C3317')
plt.legend(frameon=False, fontsize=11)
plt.grid(True, linestyle='--', alpha=0.3)
sns.despine(top=True, right=True)
plt.tight_layout()
plt.show()
```

```python
df['month'] = df['datetime'].dt.strftime('%b')

monthly_summary = df.groupby('month')[['temp', 'humidity', 'precip']].mean()

month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
monthly_summary = monthly_summary.reindex(month_order)

plt.figure(figsize=(8,5))
sns.heatmap(monthly_summary, annot=True, fmt=".1f", cmap="YlOrBr", linewidths=0.5)

plt.title("☀️ Nairobi Monthly Weather Summary (2024–2025)", fontsize=16, weight='bold', color='#5C3317', pad=15)
plt.xlabel("Weather Variable", fontsize=12, color='#5C3317')
plt.ylabel("Month", fontsize=12, color='#5C3317')
sns.despine(left=True, bottom=True)
plt.tight_layout()
plt.show()
```
